```python
mask_1.shape
```

```python
plt.figure(figsize = (10, 10))

plt.imshow(mask_1)
```

```python
mask_1.dtype
```

```python
mask_2 = torch.ones([256, 256, 3])
mask_2 = mask_2 - mask_1
```

```python
plt.figure(figsize = (10, 10))

plt.imshow(mask_2)
```
